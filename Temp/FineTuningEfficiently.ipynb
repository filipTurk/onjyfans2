```python
import torch
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, AutoTokenizer
```

```python
#model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
#gemma
#lamma 2 or 3
#mistral 7b
#XGLM
#GAMS
#"cjvt/GaMS-1B"
model_name = "cjvt/GaMS-1B"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True, # loading in 4 bit
    bnb_4bit_quant_type="nf4", # quantization type
    bnb_4bit_use_double_quant=True, # nested quantization
    bnb_4bit_compute_dtype=torch.bfloat16,
)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    trust_remote_code=True
)
model.config.use_cache = False
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token
```

```python
generator = pipeline(
    model=model, tokenizer=tokenizer,
    task='text-generation',
    temperature=0.001,
    max_new_tokens=500,
    repetition_penalty=1.1
)
```

```python
prompt = "hello do you work??"
res = generator(prompt)
print(res[0]["generated_text"])
```

```python
import random

def get_input_text(example):
    return example["messages"][0]["content"]

def get_expected_output(example):
    return example["messages"][1]["content"]

# Get 3 random examples
def get_few_shot_examples(traffic_examples, n=3):
    return random.sample(traffic_examples, n)

few_shot_examples = get_few_shot_examples(traffic_examples, n=3)

# Build the few-shot prompt
intro = """You are a traffic report generator for RTV Slovenia. Follow these rules strictly:

1. Use one of these two formulations:
   - Cesta in smer + razlog + posledica in odsek
   - Razlog + cesta in smer + posledica in odsek

2. Importance:
    - A1 very important messages
    - B1 normal messages
    - C1 unimportant messages

3. Always use proper road names and directions:
   - Example: "primorska avtocesta proti Kopru" not just "A1 proti jugu"
   - For Ljubljana bypass: specify which part (vzhodna, zahodna, severna, južna)
   - For junctions: specify both from and to directions

4. Traffic jam reporting:
   - Report only if at least 1km long
   - Include location between two points
   - Specify direction
"""

examples_str = ""
for ex in few_shot_examples:
    input_ex = get_input_text(ex)
    output_ex = get_expected_output(ex)
    examples_str += f'\nExample:\nInput: "{input_ex}"\nOutput: "{output_ex}"\n'

# Choose a new example to generate output for
new_example = random.choice(traffic_examples)
input_example = get_input_text(new_example)
expected_output = get_expected_output(new_example)

# Final prompt
prompt = f"""{intro}
{examples_str}
Now, generate a traffic report for this input:
{input_example}"""

# Generate with your model
res = generator(prompt)
text = res[0]['generated_text']

# Extract the response (optional)
start = "Now, generate a traffic report for this input:"
report_input = text.split(start, 1)[-1].strip()

# Print results
print("PROMPT:")
print(prompt)
print("\n--- GENERATED REPORT ---")
print(report_input)
print("\n--- EXPECTED OUTPUT ---")
print(expected_output)
```

```python
from datasets import load_dataset, Dataset
import json
def get_traffic_questions(test_examples) -> Dataset:
    SYSTEM_PROMPT ="You are a traffic report generator for RTV Slovenia."
    data = []
    for ex in test_examples:
        user_msg = ex['messages'][0].get('content', '')
        assistant_msg = ex['messages'][1].get('content', '')
        # Make sure both are strings
        if not isinstance(user_msg, str):
            user_msg = str(user_msg) if user_msg is not None else ''
        if not isinstance(assistant_msg, str):
            assistant_msg = str(assistant_msg) if assistant_msg is not None else ''

        data.append({
            'prompt': json.dumps([
                {'role': 'system', 'content': SYSTEM_PROMPT},
                {'role': 'user', 'content': user_msg}
            ], ensure_ascii=False),
            'answer': assistant_msg
        })
    return Dataset.from_list(data)

dataset = get_traffic_questions(traffic_examples)
print(dataset[0])
```

```python
from peft import LoraConfig, get_peft_model

lora_alpha = 32
lora_dropout = 0.1
lora_r = 16

peft_config = LoraConfig(
    lora_alpha=lora_alpha,
    lora_dropout=lora_dropout,
    r=lora_r,
    bias="none",
    task_type="CAUSAL_LM"
)
```

```python
def print_trainable_parameters(model):
    """
    Prints the number of trainable parameters in the model.
    """
    trainable_params = 0
    all_param = 0
    for _, param in model.named_parameters():
        all_param += param.numel()
        if param.requires_grad:
            trainable_params += param.numel()
    print(
        f"trainable params: {trainable_params} || all params: {all_param} || trainable%: {100 * trainable_params / all_param}"
    )
model
print_trainable_parameters(model)
```

```python
from transformers import TrainingArguments
output_dir = "./results"
per_device_train_batch_size = 1
gradient_accumulation_steps = 1
optim = "paged_adamw_32bit" #specialization of the AdamW optimizer that enables efficient learning in LoRA setting.
save_steps = 100
logging_steps = 10
learning_rate = 2e-4
max_grad_norm = 0.3
max_steps = 500
warmup_ratio = 0.03
lr_scheduler_type = "constant"
```

```python
training_arguments = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=per_device_train_batch_size,
    gradient_accumulation_steps=gradient_accumulation_steps,
    optim=optim,
    save_steps=save_steps,
    logging_steps=logging_steps,
    learning_rate=learning_rate,
    fp16=True,
    max_grad_norm=max_grad_norm,
    max_steps=max_steps,
    warmup_ratio=warmup_ratio,
    group_by_length=True,
    lr_scheduler_type=lr_scheduler_type,
    report_to="none"
)
```

```python
import json

def convert_to_text(example):
    messages = json.loads(example["prompt"])
    input_text = messages[1]["content"]  # Extract user message
    output_text = example["answer"]
    return {
        "text": f"Input: {input_text}\nOutput: {output_text}"
    }
dataset = dataset.map(convert_to_text)
```

```python
from trl import SFTTrainer

max_seq_length = 512

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_config,
    #dataset_text_field="text",
    max_seq_length=max_seq_length,
    tokenizer=tokenizer,
    args=training_arguments,
)
```

```python
trainer.train()
```

```python
model_to_save = trainer.model.module if hasattr(trainer.model, 'module') else trainer.model  # Take care of distributed/parallel training
model_to_save.save_pretrained("outputs")
```
